```python
import requests
import json
```

```python
url = 'https://api.spoonacular.com/recipes/complexSearch'
#client_id = ""
api_key = "4ef23a7b7a3640d49145346380be6f15"
```

```python
params = {
"query": 'pasta',
}

headers = {
    'Content-Type':'application/json',
    'x-api-key': api_key
}

response = requests.get(url, params=params, headers=headers)

if response.status_code == 200:
    print("Успешный запрос")
    data = json.loads(response.text)
    venues = data["results"]
    for i in range(5):
        for venue in venues:
            try:
                print("Название:", venue["title"])
                print("Изображение:", venue["image"])
                print("\n")

            except KeyError:
                print("No data")
                print("\n")
```
